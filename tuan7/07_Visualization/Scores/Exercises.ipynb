### Step 1. Import the necessary libraries

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

### Step 2. Create the DataFrame that should look like the one below.

```python
dict_data = {
    'first_name':['Jason','Molly','Tina','Jake','Amy'],
    'last_name' : ['Miller','Jacobson','Ali','Milner','Cooze'],
    'age' : [43,52,36,23,73],
    'female' : [0,1,1,0,1],
    'preTestScore':[4,24,31,2,3],
    'postTestScore':[25,94,57,62,70]
}
df = pd.DataFrame(dict_data)
%matplotlib inline
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python


# Step 1: Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*10, c='red')

# Step 2: Add labels and title
plt.title('Scatterplot of preTestScore vs. postTestScore with Point Size Determined by Age')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

# Step 3: Create a custom legend for age
ages = df['age'].unique()
for age in sorted(ages):
    plt.scatter([], [], s=age*10,  c='red', label=str(age))

plt.legend(title='Age', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
